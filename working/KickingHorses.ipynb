### 2. Kicking Horses

A famous early application of low-number statistics was an analysis of Prussian cavalryman horse-kick deaths by Bortkiewicz in 1898.

He studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (resulting in a total of 200 corps-years).  
He recorded the number of corps with a given number of deaths as follows:

| Number of Deaths | Number of Groups |
|------------------|------------------|
| 0                | 109              |
| 1                | 65               |
| 2                | 22               |
| 3                | 3                |
| 4                | 1                |

- Plot the resulting probability distribution (careful with normalization).
- How does it look like? Try to fit it by hand with a parametrized model.

(Hint: Think about the sample mean and sample variance)  


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Data
deaths = np.array([0, 1, 2, 3, 4])
groups = np.array([109, 65, 22, 3, 1])

# Normalize to get the empirical probability distribution
total_groups = np.sum(groups)
empirical_prob = groups / total_groups

# Calculate sample mean (also the lambda for Poisson)
sample_mean = np.sum(deaths * groups) / total_groups
print(f"Sample mean (λ): {sample_mean:.2f}")

# Poisson distribution with estimated lambda
poisson_probs = poisson.pmf(deaths, mu=sample_mean)

# Plotting
plt.figure(figsize=(8,5))
plt.bar(deaths - 0.2, empirical_prob, width=0.4, label="Empirical", color='skyblue')
plt.bar(deaths + 0.2, poisson_probs, width=0.4, label=f"Poisson (λ={sample_mean:.2f})", color='orange')
plt.xlabel("Number of deaths")
plt.ylabel("Probability")
plt.title("Horse-kick deaths among Prussian cavalrymen")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
